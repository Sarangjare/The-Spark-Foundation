# Importing the required librariesÂ¶


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

data_set=pd.read_csv("http://bit.ly/w-data")
print("Data read Successful")
```

First 5 rows of the data

```python
data_set.head()
```

```python
data_set.isnull().sum()
```

Plotting the scatter diagram

```python
sns.scatterplot(y=data_set["Scores"],x=data_set["Hours"])
plt.title("Marks Vs Study Hours",size=15)
plt.xlabel('Marks Percentage',size=12)
plt.ylabel('Marks Studied',size=12)
plt.show()
```

Plotting regression line

```python

sns.regplot(y=data_set["Scores"],x=data_set["Hours"])
plt.title("Marks Vs Study Hours",size=15)
plt.xlabel('Marks Percentage',size=12)
plt.ylabel('Marks Studied',size=12)
plt.show()

print(data_set.corr())
```

Prepare the data

```python
x=data_set.iloc[:,:-1].values # x=attributes
y=data_set.iloc[:,1].values # y=labels
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

Train model

```python
from sklearn.linear_model import LinearRegression

lr= LinearRegression()
lr.fit(x_train,y_train)

print('Training is completed')
```

Prediction

```python
# predict score
y_pred=lr.predict(x_test)
```

```python
# comparing actual vs predicted
df=pd.DataFrame({'Hours':[i[0] for i in x_test], "Predicted Marks":[k for k in y_pred]})
df
```

```python
compare_scores=pd.DataFrame({'Actual Marks':y_test,'Predicted Marks':y_pred})
compare_scores
```

```python

plt.scatter(x=x_test,y=y_test,color='red')
plt.plot(x_test,y_pred,color='black')
plt.title("Actual Vs Predicted",size=15)
plt.xlabel('Marks Percentage',size=12)
plt.ylabel('Marks Studied',size=12)
plt.show()
```

Evaluating model

```python
from sklearn import metrics as m
print('Mean Absolute Error:',m.mean_absolute_error(y_test,y_pred))
```
